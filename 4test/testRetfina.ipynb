```python
import cv2
import matplotlib.pyplot as plt

def draw_faces_landmarks(image_path, faces_data, output_image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Iterate over each face in the dictionary faces_data
    for face_key, face_data in faces_data.items():
        # Get the bounding box coordinates
        x1, y1, x2, y2 = face_data['facial_area']

        # Draw bounding box (rectangle) around the face
        cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Blue box

        # Draw red dots on the facial landmarks
        for key, point in face_data['landmarks'].items():
            cv2.circle(image, (int(point[0]), int(point[1])), 2, (255, 0, 0), -1)  # Red dots

    # Save the image with bounding boxes and landmarks
    cv2.imwrite(output_image_path, image)
    print(f"Image saved to {output_image_path}")



draw_faces_landmarks("c:/Users/four/Downloads/WIDER/WIDER_train/images/0--Parade/0_Parade_marchingband_1_6.jpg",resp,"out.jpg")
```
